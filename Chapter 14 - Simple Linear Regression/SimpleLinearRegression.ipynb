```python
#Try to approximate the relationship between x and y by drawing a straight line through the data.
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 7, 5, 11, 14])

# Calculate the coefficients
b, a = np.polyfit(x, y, 1)  # b is slope, a is intercept

# Print the line equation
print(f"Approximated line: y = {a:.2f} + {b:.2f}x")

# Generate y values from the line
y_pred = a + b * x

# Plot the scatter diagram
plt.scatter(x, y, color='blue', label='Data points')

# Plot the regression line
plt.plot(x, y_pred, color='red', label='Best-fit line')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot with Approximated Line')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
#Develop the estimated regression equation by computing the values of b0 and b1 using equations
# Given data
x = [1, 2, 3, 4, 5]
y = [3, 7, 5, 11, 14]
n = len(x)

# Compute the necessary sums
sum_x = sum(x)
sum_y = sum(y)
sum_x_squared = sum([xi**2 for xi in x])
sum_xy = sum([xi * yi for xi, yi in zip(x, y)])

# Compute b1 (slope)
b1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)

# Compute b0 (intercept)
x_mean = sum_x / n
y_mean = sum_y / n
b0 = y_mean - b1 * x_mean

# Print the regression equation
print(f"Estimated regression equation: y = {b0:.2f} + {b1:.2f}x")
```

```python
# Using the estimated coefficients
x_value = 4
y_pred = b0 + b1 * x_value

print(f"Predicted value of y when x = {x_value} is {y_pred:.2f}")
```

```python
Given are five observations collected in a regression study on two variables.
xi 2 6 9 13 20
yi 7 18 9 26 23
a. Develop a scatter diagram for these data.
b. Develop the estimated regression equation for these data.
c. Use the estimated regression equation to predict the value of y when x = 6.
```

```python
import matplotlib.pyplot as plt

# Given data
x = [2, 6, 9, 13, 20]
y = [7, 18, 9, 26, 23]
n = len(x)

# Step 1: Compute regression coefficients manually
sum_x = sum(x)
sum_y = sum(y)
sum_x_squared = sum([xi ** 2 for xi in x])
sum_xy = sum([xi * yi for xi, yi in zip(x, y)])

# Slope (b1)
b1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)

# Intercept (b0)
x_mean = sum_x / n
y_mean = sum_y / n
b0 = y_mean - b1 * x_mean

# Step 2: Compute predicted y values for the regression line
y_fit = [b0 + b1 * xi for xi in x]

# Step 3: Plot the scatter plot and regression line
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x, y_fit, color='red', label='Regression Line')
plt.title("Scatter Diagram with Regression Line")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()

# Step 4: Output the regression equation
print(f"Estimated regression equation: y = {b0:.2f} + {b1:.2f}x")

# Step 5: Predict y when x = 6
x_predict = 6
y_predict = b0 + b1 * x_predict
print(f"Predicted value of y when x = {x_predict} is {y_predict:.2f}")
```

5. Brawdy Plastics, Inc., produces plastic seat belt retainers for General Motors at the Brawdy
Plastics plant in Buffalo, New York. After final assembly and painting, the parts are placed
on a conveyor belt that moves the parts past a final inspection station. How fast the parts
move past the final inspection station depends upon the line speed of the conveyor belt
(feet per minute). Although faster line speeds are desirable, management is concerned that
increasing the line speed too much may not provide enough time for inspectors to identify
which parts are actually defective. To test this theory, Brawdy Plastics conducted an experiment
in which the same batch of parts, with a known number of defective parts, was
inspected using a variety of line speeds. The following data were collected.

a. Develop a scatter diagram with the line speed as the independent variable.
b. W hat does the scatter diagram developed in part (a) indicate about the relationship
between the two variables?
c. Use the least squares method to develop the estimated regression equation.

```python
import matplotlib.pyplot as plt

# Data
x = [20, 20, 30, 30, 40, 40, 50, 50]
y = [23, 21, 19, 16, 15, 17, 14, 11]

# Scatter plot
plt.scatter(x, y, color='blue')
plt.title("Scatter Diagram: Line Speed vs Defective Parts Found")
plt.xlabel("Line Speed (feet per minute)")
plt.ylabel("Number of Defective Parts Found")
plt.grid(True)
plt.show()
```

```python
# Step 1: Compute regression coefficients manually
n = len(x)
sum_x = sum(x)
sum_y = sum(y)
sum_x_squared = sum([xi ** 2 for xi in x])
sum_xy = sum([xi * yi for xi, yi in zip(x, y)])

# Slope (b1)
b1 = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)

# Intercept (b0)
x_mean = sum_x / n
y_mean = sum_y / n
b0 = y_mean - b1 * x_mean

# Print regression equation
print(f"Estimated regression equation: y = {b0:.2f} + {b1:.2f}x")
```

```python
x_predict = 25
y_predict = b0 + b1 * x_predict
print(f"Predicted number of defective parts for line speed 25: {y_predict:.2f}")
```
